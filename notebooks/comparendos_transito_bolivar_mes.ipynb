## **2. Lineas de Tendencia Mensual**

##### El presente analisis corresponde a dataset **"Comparendos Impuestos en Bolívar por Tipo de Infracción** descargado del portal nacional de datos abiertos [datos.gov.co](https://www.datos.gov.co/Transporte/Comparendos-Impuestos-en-Bol-var-por-Tipo-de-Infra/hfxy-8gt2/about_data). En este conjunto de Datos se puede visualizar los comparendos Impuestos en Bolívar por Tipo de Infracción ocurridos en el periodo 2019 al 2023, Mes a Mes.
##### A continuacion realizaremos un analisis de los datos donde realizaremos el proceso de ETL y visualizacion de los datos. Para ello, haremos uso del Lenguaje Python y sus bibliotecas incorporadas Numpy, Pandas y Matplotlib 

```python
#Importamos biliotecas necesarias para ejecutar proyecto
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
#Leemos el dataset objeto de estudio y almacenamos en una variable
archivo_comparendos_2 = (r"\Ricardo\Programacion\Data_Science\Proyectos_DS\comparendos_transito_Bolivar\data\Comparendos_Bolivar_por_Tipo_de_Infraccion.csv")
comparendo_2 = pd.read_csv(archivo_comparendos_2)
comparendo_2
```

```python
comparendo_2.info()
```

```python
comparendo_2[["FEBRERO 2019","MARZO 2019","TOTAL-19","JUNIO 2020","TOTAL 2020","TOTAL 2021","TOTAL 2022","TOTAl 2023"]] = comparendo_2[["FEBRERO 2019","MARZO 2019","TOTAL-19","JUNIO 2020","TOTAL 2020","TOTAL 2021","TOTAL 2022","TOTAl 2023"]].applymap(lambda x: x.replace("," ,""))
comparendo_2.sample(3)
```

```python
#Transformamos las columnas de tipo objeto a int64
comparendo_2[["FEBRERO 2019","MARZO 2019","TOTAL-19", "JUNIO 2020", "TOTAL 2020", "TOTAL 2021", "TOTAL 2022", "TOTAl 2023"]] = comparendo_2[["FEBRERO 2019","MARZO 2019","TOTAL-19", "JUNIO 2020", "TOTAL 2020", "TOTAL 2021", "TOTAL 2022", "TOTAl 2023"]].astype(np.int64)
comparendo_2.info()
```

```python
#Creamos variable para almmacenar las columnas con las que no queremos trabajar
comparendos_eliminar = comparendo_2[["TOTAL-19","TOTAL 2020","TOTAL 2021","TOTAL 2022","TOTAl 2023"]]
```

```python
#Eliminamos las columnas con datos anuales
comparendos_mes = comparendo_2.drop(labels=comparendos_eliminar, axis=1)
comparendos_mes.info()
```

```python
#Renombramos nombre de columna errado
comparendos_mes.rename(columns={"DIECIEMBRE 2021":"DICIEMBRE 2021"}, inplace=True)

#Aislamos las columnas de fechas para transformarlas
columnas = list(comparendos_mes.columns.drop(["INFRACCCIONES 2019", "DESCRIPCION COMP"]))
print(columnas)     
```

```python
#Transformamos formato de las etiquetas de columnas de fecha
import datetime

# Lista de meses en español
meses = ['ENERO', 'FEBRERO', 'MARZO', 'ABRIL', 'MAYO', 'JUNIO', 
         'JULIO', 'AGOSTO', 'SEPTIEMBRE', 'OCTUBRE', 'NOVIEMBRE', 'DICIEMBRE']

# Diccionario de conversión de meses
meses_dict = {mes: f"{i+1:02d}" for i, mes in enumerate(meses)}

# Convertir la lista al formato "yyyy-mm"
fechas_convertidas = [f"{fecha.split()[1]}-{meses_dict[fecha.split()[0]]}" for fecha in columnas]

print(fechas_convertidas)
```

```python
#Agregamos a la lista "fechas_convertidas" dos items eliminados anteriormente
fechas_convertidas.insert(0, "CODIGO")
fechas_convertidas.insert(1, "COMPARENDO")
print(fechas_convertidas)
```

```python
#Renombramos las columnas de DataFrame "comparendos_mes"
comparendos_mes.columns = fechas_convertidas
[comparendos_mes.columns]
```

```python
#Seleccionamos los datos 
df_comparendos_mes = comparendos_mes.query("CODIGO=='C24' | CODIGO=='B01' | CODIGO=='H02' | CODIGO=='C15' | CODIGO=='D12' | CODIGO=='C35' | CODIGO=='D02' | CODIGO=='D13' | CODIGO=='D01' | CODIGO=='C24' | CODIGO=='D07'",).reset_index(drop=True)
df_comparendos_mes.head()
```

```python
#Eliminamos columna "Codigo" 
df_comparendos_mes.drop(labels="CODIGO", axis=1, inplace=True)
df_comparendos_mes
```

## **Lineas de Tendencia de Total Comparendos por mes Periodo 2019-2023**

```python
#Transformamos DataFrame para realizar grafico de lineas
melt_df_mes = pd.melt(df_comparendos_mes, id_vars=["COMPARENDO"], var_name="MES", value_name="TOTAL")
melt_df_mes.head(5)
```

```python
#Verificamos datos del DataFrame
melt_df_mes.info()
```

```python
#Agrupamos el total de comparendos por mes y almacenamos en nueva variable
comparendo_mensual = melt_df_mes.groupby('MES').TOTAL.sum().reset_index()
comparendo_mensual.head()
```

```python
#Verificamos DataFrame transformado
comparendo_mensual.info()
```

```python
# Crear un gráfico de línea
plt.figure(figsize=(18, 6))
plt.plot(comparendo_mensual['MES'], comparendo_mensual['TOTAL'], marker='o', linestyle='-')
plt.xticks(rotation=90)

# Configurar el título y etiquetas del eje
plt.title('Total Comparendos Mensual, Bolivar, Periodo 2019-2023')
plt.xlabel('Mes')
plt.ylabel('Total')

# Mostrar el gráfico
#plt.grid(True)
plt.show()
```
